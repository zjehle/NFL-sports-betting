```python
season = {}
for year in scores_df.schedule_season.unique():
    temp_df = scores_df[scores_df['schedule_season'] == year]
    data = {}
    for index, row in temp_df.iterrows():
        h_team = temp_df.loc[index, 'team_home']
        h_score = temp_df.loc[index, 'score_home']
        a_team = temp_df.loc[index, 'team_away']
        a_score = temp_df.loc[index, 'score_away']
        if h_score > a_score:
            h_wr = 1 
            a_wr = 0
        elif a_score > h_score:
            h_wr = 0
            a_wr = 1
        else:
            h_wr = .5
            a_wr = .5
        h_margin = h_score - a_score
        a_margin = a_score - h_score
        if h_team not in data:
            data[h_team] = ([h_wr], [h_margin])
        else:
            l = len(data[h_team][0])
            wr_avg = (data[h_team][0][-1] * l + h_wr) / (l+1)
            # index 0 is win rates
            data[h_team][0].append(wr_avg)
            # index 1 is margins
            m_avg = (data[h_team][1][-1] * l + h_margin) / (l+1)
            data[h_team][1].append(m_avg)
        if a_team not in data:
            data[a_team] = ([a_wr], [a_margin])
        else:
            l = len(data[a_team][0])
            wr_avg = (data[a_team][0][-1] * l + a_wr) / (l+1)
            # index 0 is win rates
            data[a_team][0].append(wr_avg)
            # index 1 is margins
            m_avg = (data[a_team][1][-1] * l + a_margin) / (l+1)
            data[a_team][1].append(m_avg)
    season[year] = data    
```

```python
scores_df['winrate_home'] = 1 
scores_df['winrate_away'] = 1 
scores_df['margin_home'] = 1 
scores_df['margin_away'] = 1 
```

```python
season_year = 1965
for index, row in scores_df.iterrows():
    if scores_df.loc[index, 'schedule_season'] != season_year:
        teams = {}
        season_year = scores_df.loc[index, 'schedule_season']
#         print(season_year)
    names = [scores_df.loc[index, 'team_home'], scores_df.loc[index, 'team_away']]
    columns = ['home', 'away']
    for i in range(2):
        if names[i] not in teams:
            if season_year != 1966:
                if names[i] not in season[season_year-1]:
                    continue;
                scores_df.loc[index, 'winrate_' + columns[i]] = season[season_year-1][names[i]][0][-1]
                scores_df.loc[index, 'margin_' + columns[i]] = season[season_year-1][names[i]][1][-1]
                teams[names[i]] = 0
        else:
            scores_df.loc[index, 'winrate_' + columns[i]] = season[season_year][names[i]][0][teams[names[i]]]
            scores_df.loc[index, 'margin_' + columns[i]] = season[season_year][names[i]][1][teams[names[i]]]
            teams[names[i]] += 1
```

```python
#cover_favorite (initialize cover favorite as a column of all 0's)
#first step - loop through all the rows
#loop through characters of team_favorite_id and see if each are in home_team
#if all are in home_team, then we can assume home_team is team favorite
#if not then we can assume away team is team_favorite
#cover_favorite
#edge case of JAX and Pick
```

```python
scores_df['home_fave'] = True
scores_df['cover_favorite'] = 0
for index, row in scores_df.iterrows():
    if scores_df.loc[index,'team_favorite_id'] == 'PICK':
        continue
    home_is_fav = True
    if scores_df.loc[index,'team_favorite_id'] == 'JAX':
        if scores_df.loc[index,'team_home'] != 'Jacksonville Jaguars':
            home_is_fav = False
    for char in scores_df.loc[index,'team_favorite_id'].lower(): 
        if char not in scores_df.loc[index,'team_home'].lower():
            home_is_fav = False
            break
    margin = scores_df.loc[index,'score_home'] - scores_df.loc[index,'score_away'] 
    if not home_is_fav:
        scores_df.loc[index,'home_fave'] = False
        margin *= -1
    scores_df.loc[index,'cover_favorite']= margin > scores_df.loc[index, 'spread_favorite'] * -1
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
X = scores_df[['winrate_home','winrate_away','margin_home','margin_away','spread_favorite','home_fave']]
y = scores_df['cover_favorite'].astype(str)
clf = LogisticRegression(max_iter=500).fit(X, y)
scores = cross_val_score(clf, X, y, cv=10)
scores.mean()
```

```python
X = scores_df[['winrate_home','winrate_away','margin_home','margin_away', 'spread_favorite', 'home_fave']]
y = scores_df['cover_favorite'].astype(str)
clf = RandomForestClassifier().fit(X, y)
scores = cross_val_score(clf, X, y, cv=10)
scores.mean()
```
